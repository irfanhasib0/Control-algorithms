```python
import airsim
client = airsim.CarClient()
client.confirmConnection()
#client.enableApiControl(True)
#car_controls = airsim.CarControls()
client.reset()
```

```python
_map=[]
prev_state=[-1,-1]

while(1): 
    car_state=client.getCarState()
    posx=car_state.kinematics_estimated.position.x_val
    posy=car_state.kinematics_estimated.position.y_val
    if abs(prev_state[0]-posx) >1.0 or abs(prev_state[1]-posy) >1.0:
        print(posx,posy)
        _map.append([posx,posy])
        prev_state=[posx,posy]
```
