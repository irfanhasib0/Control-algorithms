```python
import numpy as np
X1=0
X2=129
Y1=-1
Y2=127
n=10
points=[[X1,Y1], [X2-n,Y1],[X2,n], [X2,Y2-n],[X2-n,Y2], [n,Y2],[X1,Y2-n], [X1,Y1]]
#points=[[0,0],[115,0],[130,15],[130,115],[115,130],[15,130],[0,115],[0,0]]
pts=np.array(points)
from matplotlib import pyplot as plt
plt.plot(pts[:,0],pts[:,1])
plt.show()
points=[]
for i in range(len(pts)-1):
    dist=np.sqrt(pts[i][0]**2+pts[i][1]**2)-np.sqrt(pts[i+1][0]**2+pts[i+1][1]**2)
    X1=pts[i,0]
    X2=pts[i+1,0]
    
    Y1=pts[i,1]
    Y2=pts[i+1,1]
    #points.append()
    if X1!=X2 :
        Xs=np.linspace(X1,X2,int(abs(X2-X1)+1))
    else :
        Xs=np.ones(int(abs(Y1-Y2)+1),dtype=np.float64)*X1
    if Y1 !=Y2 :
        Ys=np.linspace(Y1,Y2,int(abs(Y2-Y1)+1))
    else:
        Ys=np.ones(int(abs(X1-X2)+1),dtype=np.float64)*Y1
    points+=[[x,y] for x,y in zip(Xs,Ys)]
    
points=np.array(points)
plt.plot(points[:,0],points[:,1])
plt.show()
_points=points.copy()
```

```python
def moving_avg(arr):
    N=25
    _arr=arr.copy()
    for i,pt in enumerate(arr):
        if i>=N//2:
            ul=i-N//2
            ll=i+N//2
            arr[i]=np.mean(_arr[ul:ll+1])
    return arr
points[:,0]=moving_avg(points[:,0])
points[:,1]=moving_avg(points[:,1])
plt.plot(points[:,0],points[:,1])
plt.plot(_points[:,0],_points[:,1])
plt.show()
```
